```python
import imgaug.augmenters as iaa
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 공유드라이브에 올려두었습니다
img = cv2.imread("/content/drive/MyDrive/비디오 분류_1조/img1.jpg")
```

# 원본 이미지

```python
plt.imshow(img)
```

# 여러가지 증강 한번에

```python
# 이미지 증강 함수 입력
seq = iaa.Sequential([
    iaa.imgcorruptlike.GaussianNoise(severity=2),       # 노이즈
    iaa.imgcorruptlike.MotionBlur(severity=2),      # 화면 흔들림 효과
    iaa.imgcorruptlike.Snow(severity=2),        # 눈 오는 효과
    iaa.imgcorruptlike.JpegCompression(severity=2)      # 이미지 압축 깨짐
])
```

```python
generate_img = seq(image=img)
```

```python
plt.imshow(generate_img)
```

```python
#이미지를 출력할 객체

fig = plt.figure(figsize=(20, 10))

for i in range(20):
    #4줄 5칸으로 나누고 i+1번째 이미지를 그림
    subplot = fig.add_subplot(4, 5, i + 1)
    generate_img = seq(image=img)

    #subplot.set_xticks([]) : 그래프의 x축을 설정
    #데이터가 없으므로 x축에 아무 데이터도 출력 안됨
    subplot.set_xticks([])

    #subplot.set_yticks([]) : 그래프의 y축을 설정
    #데이터가 없으므로 y축에 아무 데이터도 출력 안됨
    subplot.set_yticks([])

    #증강된 이미지 출력 (X_train[0]을 회전 이동으로 데이터가 많은것 같은 효과를 줌)
    subplot.imshow(generate_img)
#그래프를 화면에 그림
plt.show()
```

# 아핀변환

```python
# 이미지 증강 함수 입력
seq = iaa.Affine(scale={"x":(0.8, 1.2), "y": (0.8, 1.2)},       # 이미지 크기를 80%~120%로 변환
                 translate_percent={"x":(-0.2, 0.2), "y":(-0.2, 0.2)},      # 이미지 이동 +-20%
                 rotate=(-25, 25),       # 각도 -25~25 사이로 회전
                 shear=(-8, 8))     # 휘어지는 효과
```

```python
generate_img = seq(image=img)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(generate_img)
```

```python
#이미지를 출력할 객체

fig = plt.figure(figsize=(20, 10))

for i in range(20):
    #4줄 5칸으로 나누고 i+1번째 이미지를 그림
    subplot = fig.add_subplot(4, 5, i + 1)
    generate_img = seq(image=img)

    #subplot.set_xticks([]) : 그래프의 x축을 설정
    #데이터가 없으므로 x축에 아무 데이터도 출력 안됨
    subplot.set_xticks([])

    #subplot.set_yticks([]) : 그래프의 y축을 설정
    #데이터가 없으므로 y축에 아무 데이터도 출력 안됨
    subplot.set_yticks([])

    #증강된 이미지 출력 (X_train[0]을 회전 이동으로 데이터가 많은것 같은 효과를 줌)
    subplot.imshow(generate_img)
#그래프를 화면에 그림
plt.show()
```
